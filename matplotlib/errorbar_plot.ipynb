```python
#!/usr/bin/env python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# example data
x = [20,38,75,187,380,615]
y = [19.4,8,4.9,3.9,3.6,3.4]
yerr = [1.1,1,0.7,0.6,0.35,0.3]

plt.errorbar(x, y, yerr=yerr, fmt = "o", capsize = 5, capthick = 1)
plt.xlabel('Drift field [V/cm]')
plt.ylabel('Mean correction [%]')
plt.grid(True)

fig = plt.gcf()
plt.show()
plt.draw()
fig.savefig('plot0.png', dpi=300)
```
